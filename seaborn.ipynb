## Statistical Data visualization using Seaborn

```python
import seaborn as sbn
from scipy import stats
import numpy as np
import pandas as pd
from scipy.stats import spearmanr
```

## Distplot

```python
tips = sbn.load_dataset('tips')
```

```python
#shows histogram and kde - kernal density estimation for the data 
sbn.distplot(tips['total_bill'])
```

```python
# hist - histogram, kde - kernal density estimation, for large dataset kde and rug are costly
sbn.distplot(tips['total_bill'], rug = True, hist = False, kde = True)
```

```python
sbn.distplot(tips['total_bill'], rug = False, hist = True, kde = False, norm_hist=False)
```

```python
# using fit
sbn.distplot(tips['total_bill'],fit = stats.gamma,kde = False)
```

## Strip plot

```python
tips = sbn.load_dataset('tips')
```

```python
#default strip plot
sbn.stripplot(tips['total_bill'])
```

```python
# more than one axis
sbn.stripplot(x = tips['day'], y = tips['total_bill'])
```

```python
# other methods
sbn.stripplot(x = 'day', y = 'total_bill', data = tips)
```

```python
# jitter
sbn.stripplot(x = 'day', y = 'total_bill', data = tips, jitter=False)
```

```python
# jitter
sbn.stripplot(x = 'day', y = 'total_bill', data = tips, jitter=0.5)
```

```python
# hue
sbn.stripplot(x = 'sex', y = 'total_bill',hue = 'day', data = tips, jitter=True)
```

```python
# split = false
sbn.stripplot(x = 'day', y = 'total_bill',hue = 'smoker', data = tips, jitter=True, split = False)
```

```python
# split = true
sbn.stripplot(x = 'day', y = 'total_bill',hue = 'smoker', data = tips, jitter=True, split = True)
```

```python
# color
sbn.stripplot(x = 'tip', y = 'day',hue = 'smoker', data = tips,  color = ".3")
```

```python
# strip and box plot
sbn.boxplot(x = 'tip', y = 'day', data = tips,  whis= np.inf)
sbn.stripplot(x = 'tip', y = 'day', data = tips,  color = ".3")
```

```python
# swarm plot
sbn.swarmplot(x = 'tip', y = 'day',hue = 'smoker', data = tips, palette='Set2', split = True)
```

Difference between strip plot and swarm plot:

A swarm plot is very similar to a strip plot, yet the locations of points are adjusted automatically to avoid overlap even if the jitter value is not applied.

## Joint plot

```python
sbn.jointplot(x = 'total_bill', y = 'tip', data=tips)
```

```python
# kind  = reg, regression
sbn.jointplot(x = 'total_bill', y = 'tip', data=tips, kind= 'reg')
```

```python
# kind  = hex, hexagon shaped heatmap/histogram
sbn.jointplot(x = 'total_bill', y = 'tip', data=tips, kind= 'hex')
```

```python
# kind = kde
sbn.jointplot(x = 'total_bill', y = 'tip', data=tips, kind= 'kde')
```

```python
# spearmanr
sbn.jointplot(x = 'size', y = 'total_bill', data=tips, stat_func= spearmanr)
```

```python
# join more plots
sbn.jointplot(x = 'total_bill', y = 'tip', data=tips).plot_joint(sbn.kdeplot, n_levels = 6)
```

```python
# join more plots
sbn.jointplot(x = 'total_bill', y = 'tip', data=tips).plot_joint(sbn.kdeplot, n_levels = 6).plot_marginals(sbn.rugplot)
```
